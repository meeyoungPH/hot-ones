```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import pprint as pp
```

```python
# setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of data source
url = "https://www.saucemania.com.au/hot-sauce/a-complete-list-of-hot-ones-hot-sauces/"

# open URL in instance of Chrome
browser.visit(url)
```

```python
html = browser.html

# create BeautifulSoup object and parse
soup = bs(html, 'lxml')
```

```python
entries = soup.find('div', class_='entry-content')

sauces = []
season_count = 1
for i in range(1,23):
    try: 
        season = entries.find('table', id='tablepress-'+ str(i))
    
        for j in range(2,12):
            row = season.find('tr', class_='row-' + str(j))
            
            try:
                url = row.find('a', href=True)['href']
            except:
                url = ""
            
            dict = {
                'season': season_count,
                'order': row.find('td', class_='column-1').text,
                'sauce': row.find('td', class_='column-2').text,
                'url': url,
                'scoville_rating' : row.find('td', class_='column-3').text
            }
            sauces.append(dict)
        season_count += 1
    except:
        # sauces.append("boop")
        next
print(pp.pprint(sauces))
```

```python
sauces_df = pd.DataFrame(sauces)
sauces_df
```
